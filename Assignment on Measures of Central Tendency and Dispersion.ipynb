```python
#5The management needs your help to give bonus points to the top 25% of students based on their math score, so how will you help the management to achieve this
```

```python
s_data=data.sort_values(by=["math score"], ascending=False)
s_data
```

```python
q75 =s_data['math score'].quantile(q=0.75)
s_data[s_data['math score'].ge(q75)]
```

Here we got the Top 25% of students based on their math score
